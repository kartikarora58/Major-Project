```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
import pickle
```

```python
parent_dir='E:/Major Project/wheat/train'
```

```python
dirlist=os.listdir(parent_dir)
```

```python
labels=dirlist[:]
```

```python
labels = sorted(labels)
labels
```

```python
data=[]
i=0
```

```python
for item in labels:
    i=labels.index(item)
    for image_name in os.listdir(os.path.join(parent_dir,item)):
        print(image_name)
        image=cv2.imread(f"{parent_dir}/{item}/{image_name}")
        image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY )
        image=image/255
        image=cv2.resize(image,(128,128))
        image=image.flatten(order='C')
        print(len(image))
        data.append([image,i])
```

```python
print(len(data))
```

```python
print(data)
```

```python
pick_in=open('wheat.pickle','wb') 
pickle.dump(data,pick_in)
pick_in.close()
```

```python
pick_in=open('wheat.pickle','rb')
data=pickle.load(pick_in)
pick_in.close()
```

```python
data
```

```python
features=[]
targets=[]
```

```python
for item1,item2 in data:
    features.append(item1)
    targets.append(item2)
```

```python
len(features)
```

```python
len(targets)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
np.array(features).shape
```

```python
model.fit(features,targets)
```

```python
pick_in=open('wheat_test.pickle','rb')
data=pickle.load(pick_in)
pick_in.close()
```

```python
X_test=[]
y_test=[]
for item1,item2 in data:
    X_test.append(item1)
    y_test.append(item2)
```

```python
print(len(y_test))
```

```python
model.score(X_test,y_test)
```
